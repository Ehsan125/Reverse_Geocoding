```python
AQ_data["geom"]= AQ_data["Y"].map(str)+','+AQ_data["X"].map(str)
```

```python
AQ_data["geom"][0]
```

```python
locator=Nominatim(user_agent="myGeocoder",timeout=10)
```

```python
rgeocode=RateLimiter(locator.reverse, min_delay_seconds=0.001)
```

```python
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
AQ_data['address']=AQ_data['geom'].progress_apply(rgeocode)
```
